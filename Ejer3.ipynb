```python
#ejercicio 3
#Perceptron multicapa para funcion XOR

from keras.models import Sequential
from keras.layers.core import Dense
import numpy as np

#patrones de entrada XOR
entrada = np.array([[1,3],[1,5],[2,6],[3,6],[1,0],[1,1],[4,1],[5,1]]) #0,1,1,0

salida = np.array([[-1],[-1],[-1],[-1],[1],[1],[1],[1]])

#costruir modelo de red neuronal multicapa
modelo = Sequential()

#definamos el numero de neuronas la capa de entrada
# el primer numero despues de "Dense" es el numero de neuronas en la capa oculta 
modelo.add(Dense(8,input_dim=2,kernel_initializer="uniform",activation="relu"))

# Segunda capa oculta 
modelo.add(Dense(8,kernel_initializer="uniform",activation="relu"))

#capa de salida (una neurona en la capa de salida)
modelo.add(Dense(2,activation="sigmoid"))

#Establecer parametros de  entrenamiento
modelo.compile(loss='mean_squared_error',optimizer='adam',metrics=['binary_accuracy'])

#Entrenamiento 
modelo.fit(entrada,salida,epochs=10)

#Score
score = modelo.evaluate(entrada,salida)
#print(score[0])

predicciones = modelo.predict(entrada)
print(predicciones.round())
```

```python
#modificado2
import random as rnd

def signo(x):
  """
  Función escalón: devuelve 1 si x es mayor o igual a cero, -1 en caso contrario.
  """
  if x >= 0:
    return 1
  else:
    return -1

def inicializar_parametros():
  """
  Inicializa los pesos y el umbral con valores aleatorios entre -0.5 y 0.5.
  """
  pesos = [rnd.uniform(-0.5, 0.5), rnd.uniform(-0.5, 0.5)]
  umbral = rnd.uniform(-0.5, 0.5)
  return pesos, umbral

if __name__ == "__main__":
  # Patrones de entrada y sus respectivas salidas esperadas
  patrones = [
    [1, 3, -1],
    [1, 5, -1],
    [2, 6, -1],
    [3, 6, -1],
    [1, 0, 1],
    [1, 1, 1],
    [4, 1, 1],
    [5, 1, 1]
  ]

  # Inicialización de los pesos y el umbral
  pesos, umbral = inicializar_parametros()
  print("Pesos iniciales:", pesos)
  print("Umbral inicial:", umbral)

  # Entrenamiento del perceptrón
  for patron in patrones:
    # Cálculo de la salida del perceptrón para el patrón actual
    entrada = patron[:2]
    salida_esperada = patron[2]
    salida = signo(entrada[0] * pesos[0] + entrada[1] * pesos[1] - umbral)
    
    # Actualización de los pesos y el umbral si la salida obtenida es distinta a la esperada
    if salida != salida_esperada:
      error = salida_esperada - salida
      pesos[0] += error * entrada[0]
      pesos[1] += error * entrada[1]
      umbral -= error
      
  print("Pesos finales:", pesos)
  print("Umbral final:", umbral)
```
